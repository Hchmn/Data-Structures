```python
class Node:
    def __init__(self,value):
        self.left = None
        self.right = None
        self.value = value
        self.parent = None
    
    def set_right(self, right):
        self.right = right
    
    def set_left(self, left):
        self.left = left
    
    def get_right(self):
        return self.right
    
    def get_left(self):
        return self.left
    
    def get_value(self):
        return self.value

class SearchTree:
    def __init__(self):
        self.root = None
    
    def insert(self, Nodeval): 
    
        if self.root == None:
            node = Node(Nodeval) 
            self.root = node
        
        else:
            self.traverse_insert(Nodeval,self.root)
   
    def traverse_insert(self, Nodeval,Root):
        if Nodeval < Root.get_value():
            if Root.get_left() == None:
                node = Node(Nodeval)
                Root.set_left(node)
            else:
                return self.traverse_insert(Nodeval, Root.get_left())
                
        else: 
            if Root.get_right() == None:
                node= Node(Nodeval)
                Root.set_right(node)
            else:
                return self.traverse_insert(Nodeval, Root.get_right())
    
    def delete(self, Nodeval):
        if self.root.get_right()== None and self.root.get_left()==None:
            self.root==None
        else:
            self.traverse_delete(Nodeval, self.root)
            
    def traverse_delete(self, Nodeval, Root):
        if Nodeval < Root.get_value():
            if Root.get_left().get_value() == Nodeval: 
                Root.set_left(None)
            else:
                return self.traverse_delete(Nodeval, Root.get_left())
                
        else: 
            if Root.get_right().get_value() == Nodeval:
                Root.set_right(None)
            else:
                return self.traverse_delete(Nodeval, Root.get_right())
            
    def traverse_preorder(self, root):
        
        if root!=None:
            print(root.get_value())
            self.traverse_preorder(root.get_left())
            self.traverse_preorder(root.get_right())
    
    def preorder(self):
        self.traverse_preorder(self.root)
            
search = SearchTree()
search.insert(100)
search.insert(7)
search.insert(1)
search.insert(-1)
search.insert(3)
search.delete(-1)
search.preorder()
```
