```python
import numpy as np
import time
import threading
from random import randint
class Stack:
    def __init__(self):
        self.top = np.NaN
        self.no_of_nodes = 0
    
    def push(self, Node):
        if self.no_of_nodes ==0:
            self.top = Node
            self.no_of_nodes+=1
            print("pushed ",self.top.get_value())
            print("no of nodes " + str(self.no_of_nodes))
        else:
            Node.set_next(self.top)
            self.top.set_prev(Node)
            self.top=Node
            self.no_of_nodes+=1
            print("pushed " + str(self.top.get_value()))
            print("no of nodes " + str(self.no_of_nodes))
    
    def pop(self):
        if self.no_of_nodes!=0:
            print("popped " + str(self.top.get_value()))
            self.temp_node = self.top.get_next()
            self.top=self.temp_node
            self.no_of_nodes-=1
    
    def peek(self):
        print("top is " + str(self.top.get_value()))
        return self.top.get_value()

class Node:
    def __init__(self, val):
        self.value = val
        self.prev_value = np.NaN
        self.next_value = np.NaN
    
    def get_next(self):
        return self.next_value
    
    def set_prev(self, Node):
        self.prev_value= Node
    
    def set_next(self, Node):
        self.next_value =Node
   
    def get_value(self):
        return self.value
    
    def get_prev(self):
        return self.prev_value

Stack1 = Stack()
node1 =Node(1)
node2= Node(2)
node3 =Node(3)
Stack1.push(node1)
Stack1.push(node2)
Stack1.push(node3)
Stack1.peek()
Stack1.pop()
Stack1.peek()
Stack1.pop()
Stack1.peek()
        
```
