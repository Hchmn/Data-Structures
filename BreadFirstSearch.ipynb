```python
:class Vertex:
    
    def __init__(self, value):
        self.value = value
        self.degrees = 0
        self.neighbors = []
        self.traversed = False
        
    def get_neighbors(self):
        return self.neighbors
    
    def get_value(self):
        return self.value
    
    def get_degree(self):
        return self.degrees
    
    def set_neighbors(self, ran1):
        if ran1 not in self.neighbors and self != ran1:
            self.neighbors.append(ran1)
            self.degrees += 1
    
    def set_traverse(self, traversed):
        self.traversed = traversed

    def get_traverse(self):
        return self.traversed


# In[13]:


import numpy as np
import random

class Graph:
    def __init__(self):
        self.vertices = []
        self.ranlist = list(np.arange(1,100))
        
    def get_vertices(self):
        return self.vertices
        
    def set_vertices(self, vertex):
        self.vertices.append(vertex)
    
    def generate(self):
        for i in range (0,30):
            random_ = random.choice(self.ranlist)
            value = Vertex(random_)
            self.set_vertices(value)
            self.ranlist.remove(random_)
            for i in range (0,len(self.get_vertices())):
                ran1 = random.choice(self.get_vertices())
                ran1.set_neighbors(value)
                value.set_neighbors(ran1)
        
    def traversal(self):
        first_vertex = random.choice(self.vertices)
        print('Breadth First Search: ', end='[ ')
        self.traverse(first_vertex)
        print(end=']')

    def traverse(self, start):
        print(start.get_value(), end=' ')
        start.set_traverse(True)
        self._traverse(start)

    def _traverse(self, v):
        already_traversed = True
        for vertex in v.get_neighbors():
            if not vertex.get_traverse():
                print(vertex.get_value(), end=' ')
                vertex.set_traverse(True)
                already_traversed = False
        if already_traversed:
            return
        for vertex in v.get_neighbors():
            self._traverse(vertex)


g = Graph()
g.generate()
for i in g.get_vertices():
        print(i.get_value())
        print("\nNeighbors:")
        for u in i.get_neighbors():
            print(u.get_value())
        print()


# In[15]:


g.traversal()
```
